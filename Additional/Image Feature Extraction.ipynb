```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
from skimage.io import imread, imshow
from skimage.filters import prewitt_h,prewitt_v
image = imread('input.png', as_gray=True)
imshow(image)
```

```python
#checking image shape 
image.shape, image
```

```python
#pixel features

features = np.reshape(image, (800*800))

features.shape, features
```

```python
image.shape
```

```python
#convert image to rgb
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb.shape
```

```python
feature_matrix = np.zeros((800,800)) 
feature_matrix.shape
```

```python
for i in range(0,image_rgb.shape[0]):
    for j in range(0,image_rgb.shape[1]):
        feature_matrix[i][j] = ((int(image_rgb[i,j,0]) + int(image_rgb[i,j,1]) + int(image_rgb[i,j,2]))/3)
```

```python
features = np.reshape(feature_matrix, (800*800)) 
features.shape
```

```python

im = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
edges_prewitt_horizontal = prewitt_h(im)
#calculating vertical edges using prewitt kernel
edges_prewitt_vertical = prewitt_v(im)

imshow(edges_prewitt_vertical, cmap='gray')
```
